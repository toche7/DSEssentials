## เตรียมข้อมูล

```python
X = df[["Passenger"]]
type(X)
```

```python
df.shape
```

```python
y = df.Fuel_Cost
y.shape
```

## plot graph of data

```python
# prompt: plot X vs y as scatter

import matplotlib.pyplot as plt

plt.scatter(X, y)
plt.xlabel('Passenger')
plt.ylabel('Fuel_Cost')
plt.title('Passenger vs Fuel Cost')
plt.show()
```

## สร้าง Model Object

```python
from sklearn.linear_model import LinearRegression
```

```python
modelObj = LinearRegression()
```

## การฝึกแบบจำลอง

```python
modelObj.fit(X, y)
```

y = theta0 + theta1*X

```python
theta1 = modelObj.coef_
```

```python
theta0 = modelObj.intercept_
```

```python
theta0 + theta1*(100)
```

## การทำนายค่า

```python
modelObj.predict([[100]])
```

```python
Xtest = pd.DataFrame([[60],[70],[80], [90], [100]])
Xtest
```

```python
Xtest.columns = ['Passenger']
```

```python
Xtest
```

```python
Ytest = modelObj.predict(Xtest)
```

```python
# prompt: plot X vs y as scatter plot, plot Xtest vs Ytest  as line plot

import matplotlib.pyplot as plt
plt.scatter(X, y, label='Training Data')
plt.plot(Xtest, Ytest, color='red', label='Prediction')
plt.xlabel('Passenger')
plt.ylabel('Fuel_Cost')
plt.title('Passenger vs Fuel Cost')
plt.legend()
plt.show()
```

## การวัดประสิทธิภาพ

### R^2

```python
modelObj.score(X,y)
```

### RMSE

```python
yPredit = modelObj.predict(X)
```

```python
y
```

```python
np.sqrt(np.mean((yPredit - y)**2))
```

```python
from sklearn.metrics import root_mean_squared_error
```

```python
root_mean_squared_error(y, yPredit)
```
