# Lab 2: Linear Regression

Linear Regression Model


$$y = \beta_0 + \beta_1x_1 \tag{1} $$



```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel('https://github.com/toche7/DataSets/raw/main/cost.xlsx')
```

```python
df
```

```python
import matplotlib.pyplot as plt
plt.scatter(df.Passenger, df['Fuel Cost'], c ='r')
plt.xlabel('X')
plt.ylabel('Y')
```

## เตรียมข้อมูล (Data Preparation)

```python
X = df[['Passenger']]
y = df['Fuel Cost']

X.shape
```

```python
y.shape
```

## Create Model Object

```python
from sklearn.linear_model import LinearRegression
```

```python
modelObj = LinearRegression()
```

## Train Model

```python
modelObj.fit(X, y)
```

## Model Evaluation

Model Performance  คำนวณจากค่า $R^2$

```python
modelObj.score(X, y)
```

```python
modelObj.coef_
```

```python
modelObj.intercept_
```

## Model Prediction

```python
b1 = modelObj.coef_ # beta1
```

```python
b0 = modelObj.intercept_ # beta0
```

```python
yhat = b0 + b1*X
```

```python
yhat
```

หรือเราสามารถใช้คำสั่งสำเร็จรูปได้เลยดังนี้

```python
modelObj.predict(X)
```

```python
import matplotlib.pyplot as plt
plt.scatter(df.Passenger, df['Fuel Cost'], c ='r')
plt.plot(X, yhat, c='b')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(['yhat', 'y'])
```
